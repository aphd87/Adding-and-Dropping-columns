```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML
transactions = pd.read_excel("grocery_database.xlsx", sheet_name = "transactions")
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
transactions
```

```python
#New column
transactions["store_id"]=1 #All observations have 1
```

```python
transactions
```

```python
#Profit column
transactions["profit"] = transactions["sales_cost"]*0.2
```

```python
transactions
```

```python
transactions["sales_type"] = np.where(transactions["sales_cost"] > 20, "Large", "Small")
#Apply a where statement
#Where sales cost are greater than 20 should be Large, otherwise Small
```

```python
transactions
```

```python
#Select funcitonality
#List of conditions that we want to apply
condition_rules = [transactions["sales_cost"]>50, transactions["sales_cost"]>20, transactions["sales_cost"]>10]
#Outcomes associated with this list
outcomes = ["X-Large", "Large","Medium"]
#Now combine both
transactions["sales_type"] = np.select(condition_rules, outcomes, default = "Small") #Small applies to anything that has not been labeled already
```

```python
transactions
#We can see medium, small, and large here.
```

```python
#Remove columns
new_df = transactions.drop(["sales_cost"], axis=1) #axis=1 drops a column, for rows use 0
```
